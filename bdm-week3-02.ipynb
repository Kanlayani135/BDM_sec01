### EX2.  Exploration of Sex and Survived using Bar Chart

```python
import numpy as np
titanics = np.genfromtxt('../input/titanic/titanic.txt',delimiter='\t',missing_values = None,dtype=str)
```

```python
titanics[0,:]
```

```python
sex = np.unique(titanics[1:,2])
sex
```

```python
sex_value = [sum(titanics[:,2]==sex[0]),
            sum(titanics[:,2]==sex[1])]
sex_value
```

```python
survived =np.unique(titanics[1:,11])
survived
```

```python
survived_value = [sum(titanics[:,11]==survived[0]),
                 sum(titanics[:,11]==survived[1])]
survived_value
```

```python
sub1 = plt.subplot(1,2,1)
sub1 = plt.bar(sex,sex_value)
sub2 = plt.subplot(1,2,2)
sub2 = plt.bar(survived,survived_value)
```
