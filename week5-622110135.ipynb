```python
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/Clinic/master/ultima_all_clean.csv', sep=',', header = 0)
df
```

### 1.Set index of data to hn numbers

```python
df = df.set_index('hn')
df
```

### 2.Remove missing values

```python
df.isnull()
```

```python
df_missing = df
df_missing = df_missing.dropna(0,'any')
df_missing
```

### 3.Remove these unnecessary columns hiv, size, utmet, vgmet, surgery, pchemo

```python
df_uncessary = df_missing.drop(columns=['hiv', 'size','utmet','vgmet', 'surgery', 'pchemo'], axis=1)
df_uncessary
```

### 4.Change age into 2 values 'Young' (age <= 20) and 'Old' (age > 20)

```python
df_uncessary['age'].unique()
```

```python
df_change_age = df_uncessary
df_change_age.loc[df_change_age['age'] <= 20,'age']  = 'Young'
df_change_age.loc[df_change_age['age'] > 20,'age'] = 'Old'
df_change_age
```

### 5.Separate data into groups based on pmmet values

```python
df_uncessary['pmmet'].unique()
```

```python
df_pm0 = df_change_age.loc[df_change_age['pmmet'] == 0.0]
df_pm0
```

```python
df_pm1 = df_change_age.loc[df_change_age['pmmet'] == 1.0]
df_pm1
```

### 6.for each group, randomly choose 10 data
