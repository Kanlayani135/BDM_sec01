#### 2.Write a Python program to count the number of even and odd numbers from a series of numbers.

```python
sample_num = [1,2,3,4,5,6,7,8,9]
odd, even = [i for i in sample_num if i %2==1], [i for i in sample_num if i %2==0]
print("Number of odd numbers:" + str(len(odd))+ "\n" + "Number of even numbers:" + str(len(even)))
```
